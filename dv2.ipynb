```python
x=np.linspace(0,10,100)
y=np.linspace(0,10,100)
#plt.plot(x,y)
fig,ax=plt.subplots()
ax.plot(x,x**2)
```

```python
x=np.linspace(5,10,100)

#plt.plot(x,y)
fig,ax=plt.subplots(nrows=2,ncols=2)
ax[0][0].plot(x,x**2,label="quadractic",color="green")
ax[0][0].legend()
ax[0][1].plot(x,x**3,label="cubic",color="orange")
ax[1][0].plot(x,x**4,label="4 th power",color="blue")
ax[1][1].plot(x,x,label="linear",color="yellow")
fig.tight_layout()
```

```python
y=np.linspace(0,10,100)
fig,ax=plt.subplots()
ax.scatter(y,np.sin(y),color="purple")
```
