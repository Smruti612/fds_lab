```python
#one hot coding with pandas , so encode brands with is owners

# we use get_dummies functions for it
pd.get_dummies(d1,columns=['fuel','owner'])
```

OneHotEncoding using Sklearn

```python
from sklearn.model_selection import train_test_split
```

```python
x=d1.iloc[:,0:4]
y=d1.iloc[:,-1]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
t1=ColumnTransformer(transformers=[("tnf2",OneHotEncoder(sparse=False,drop="first"),["fuel","owner"])],remainder="passthrough")
```

```python
df2=t1.fit_transform(x_train)
```

```python
pd.DataFrame(df2)
```

```python
df3=t1.fit_transform(x_test)
df3
```
