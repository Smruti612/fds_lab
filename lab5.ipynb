Write a Python program to plot two or more lines with legends, different widths and colors.

```python
x=[5,10,15,20,25,30]
y1=[1,2,3,4,5,6]
y2=[3,6,9,12,15,18]
y3=[6,12,18,24,30,32]

plt.plot(x,y1,label='lip color',color='blue', linewidth=2)
plt.plot(x,y2,label='eyeshadow',color='green', linewidth=5.5)
plt.plot(x,y3,label='blush',color='purple', linewidth=8)
plt.title('Consumer graphs')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

  Write a Python program to plot two or more lines and set the line markers.

```python
x=[1,2,3,4]
y4=[10,20,30,40]
y3=[7,14,21,28]
x3=[8,16,24,32]
plt.plot(x,y4,marker='D',markersize=5, markerfacecolor='yellow')
plt.plot(x3,y3,marker='o',markersize=10, markerfacecolor='green')
plt.plot(x3,y4,marker='x',linestyle='--',markersize=8, markerfacecolor='blue')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Markers')

plt.show()
```

Write a Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016.
Sample Financial data (fdata.csv):


```python
f1=pd.read_csv("./files/sample.csv")
f1
```

```python
x_pos=f1.iloc[ : ,0]
x_pos
```

```python
y_pos1=f1.iloc[:,1]
y_pos2=f1.iloc[:,2]
y_pos3=f1.iloc[:,3]
y_pos4=f1.iloc[:,4]
y_pos1
```

```python
fig,ax=plt.subplots(figsize=(5,5))
ax.plot(x_pos,y_pos1,label='Open')
ax.legend()
ax.plot(x_pos,y_pos2,label='High')
ax.legend()
ax.plot(x_pos,y_pos3,label='Low')
ax.legend()
ax.plot(x_pos,y_pos4,label='Close')
ax.legend()
plt.title('financial data of Alphabet Inc.')
plt.xlabel('X-axis,Date ')
fig.tight_layout()
```
