```python
t1=np.arange(0,5,0.1)
plt.figure()
plt.subplot(2,1,1) # so 2,1,1 means 2 rows and 1 column 
plt.plot(t1,t1**2,'*')
plt.subplot(2,1,2) # so 2,1,1 means 2 rows and 1 column 
plt.plot(t1,t1**4,'*')
```

```python
x=np.linspace(start=-5,stop=5,num=150)
fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0,0].set_title('Upper left')
axes[0,1].set_title('Upper right')
axes[1,0].set_title('lower left')
axes[1,1].set_title('lower right')

axes[0,0].plot(x,x,label='Linear')
axes[0,0].legend()
axes[0,1].plot(x,x**2,label='quadractic',color='green')
axes[0,1].legend()
axes[1,0].plot(x,x**3,label='cubic',color='red')
axes[1,0].legend()
axes[1,1].plot(x,x**4,label='4th power',color='purple')
axes[1,1].legend()
fig.tight_layout()
```

```python
x=np.linspace(start=-5,stop=5,num=150)
fig,axes=plt.subplots(figsize=(6,6))

axes.plot(x,x,label='Linear')
axes.legend()
axes.plot(x,x**2,label='quadractic',color='green')
axes.legend()
axes.plot(x,x**3,label='cubic',color='red')
axes.legend()
axes.plot(x,x**4,label='4th power',color='purple')
axes.legend()
fig.tight_layout()
```
