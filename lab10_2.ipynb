**Does BMI of males differ significantly from that of females?**
-  **Step 1**: State the null and alternative hypothesis
- **Null Hypothesis**: proportions of fraud cases in rural and urban area are equal.
- **Alternative Hypothesis**: proportions of fraud cases in rural and urban area are not equal.

- **Step 2**:Select an appropriate statistical test and the corresponding test statistic
- We select proportion_ztest 
- **Step 3**: Choose level of significance
- alpha = 0.05

- **Step 4** : Collection of data and calculate value of test case

```python
rural_fraud_cases = d1[d1["AccidentArea"] == "Rural"]["FraudFound_P"]
urban_fraud_cases = d1[d1["AccidentArea"] == "Urban"]["FraudFound_P"]
```

```python
count_fraud=[rural_fraud_cases.sum(),urban_fraud_cases.sum()]
total_no=[len(rural_fraud_cases),len(urban_fraud_cases)]
```

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
t_statistic, p_value = proportions_ztest(count_fraud,total_no)
print(f't_statistic: {t_statistic}\np_value: {p_value}')
```

```python
#comparing value with alpha 
alpha=0.05
p_value > alpha
```

- **Since, we have got false as result , we conclude that WE REJECT THE NULL HYPOTHESIS as there is enough evidence that fraud cases are different in urban ad rural acciden area.**
