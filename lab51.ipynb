```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
import pandas as pd
```

```python
def function(x,y):
    return np.sin(np.sqrt(x**2+y**2))
```

```python
x=np.linspace(-10,10,40)
y=np.linspace(-10,10,40)
```

```python
X,Y=np.meshgrid(x,y)
Z=function(X,Y)
```

```python
fig=plt.figure(figsize=(10,8))
ax=plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,cmap='cool',alpha=0.8)
ax.set_title('3D Contour Plot of function(x, y) =\sin(sqrt(x^2 + y^2))', fontsize=14)
ax.set_xlabel('x',fontsize=12)
ax.set_ylabel('y',fontsize=12)
ax.set_zlabel('z',fontsize=12)
plt.show()
```

 Write a Python programming to create a pie chart of the popularity of programming Languages.
Sample data:use sample2.csv

```python

df=pd.DataFrame(
    {
        "languages":['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'],
        "popularity":[22.2,17.6,8.8,8,7.7,6.7]
    }
)
df
```

```python
plt.figure(figsize=(5,5))
plt.pie(df['popularity'],labels=df['languages'],shadow=False,autopct="%1.1f%%")
```

Write a Python program to draw a scatter plot for three different groups comparing weights and heights.

```python
height=[[160, 165, 170, 175, 180],
           [155, 160, 165, 170, 175],
           [150, 155, 160, 165, 170]]

weight=[[60, 65, 70, 75, 80],
           [55, 60, 65, 70, 75],
           [50, 55, 60, 65, 70]]

plt.figure(figsize=(6,6))
plt.scatter(height[0],weight[0],color="purple",label="Group A",marker="o")
plt.scatter(height[1],weight[1],color="red",label="Group B",marker="x")
plt.scatter(height[2],weight[2],color="green",label="Group C",marker="D")
plt.title("Height(cm) vs Weight(kg)")
plt.xlabel("height")
plt.ylabel("Weight")
plt.legend()
plt.show()
```
